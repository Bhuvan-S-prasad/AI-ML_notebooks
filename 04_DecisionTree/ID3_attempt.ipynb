```python
def entropy_target(attribute):
  attribute_unique = attribute.unique()
  attribute_value_counts = attribute.value_counts()

  div = attribute_value_counts / len(attribute)

  entropy = -np.sum( div * np.log2(div))

  return entropy
```

```python

def x_columns_unique_values_in_target_count(x, y):
  y_unique = y.unique()

  for attribute_name in x.columns:
    print(f"\nCounts for '{attribute_name}':")
    attribute_unique = x[attribute_name].unique()

    for att_value in attribute_unique:
      for target_value in y_unique:
        count = len(df[(df[attribute_name] == att_value) & (df['Job Offer'] == target_value)])
        print(f"  Attribute value: '{att_value}', Target value: '{target_value}', Count: {count}")

```

```python
total = 0
attributes_list = []
target_list = []
count_list = []
total_list = []

for col in x.columns:                     #col : cgps, interactive etc
  attribute_unique = x[col].unique()      #attribute_unique : [ >=9, >=8, <8]

  for att in attribute_unique:            #>=9 each unique value
    for target in y_unique:               #target : [yes, no]
      count = len(df[(df[col] == att) & (df['Job Offer'] == target)])
      total = total + count
      print(f"attribute: {att},  target: {target}, count: {count}, total: {total}")
      attributes_list.append(att)
      target_list.append(target)
      count_list.append(count)
      total_list.append(total)


    total = 0






```

```python
entropy_info = {'attributes': attributes_list, 'target': target_list, 'count': count_list, 'total': total_list}
entropy_df = pd.DataFrame(entropy_info)
```
