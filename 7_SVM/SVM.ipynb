```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drug.csv')
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['Drug'])
plt.title("countplot for Drug")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(df['Drug'])
plt.title("countplot for Drug")
plt.show()
```

```python
df.hist(bins=30, figsize=(12,6), edgecolor='black', color='crimson')
plt.suptitle("histogram of all numerical features")
plt.show()
```

```python
sns.pairplot(df, diag_kind='kde')
plt.show()
```

```python
plt.figure(figsize=(12,8))

for i, col in enumerate(df.columns):
  plt.subplot(3,3,i+1)
  sns.boxplot(y=df[col])
  plt.tight_layout()
  plt.plot()
```

```python
correlation_mat = df.corr
```

```python
numerical_df = df.select_dtypes(include=np.number)

correlation_mat = numerical_df.corr()
plt.figure(figsize=(12,8))
sns.heatmap(correlation_mat, annot=True, fmt='.2f', cmap='coolwarm')
plt.title("correlation matrix of numerical features")
plt.show()
```
