```python
from langchain_chroma import Chroma
from langchain_core.prompts import PromptTemplate
from langchain_text_splitters import CharacterTextSplitter
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
```

```python
with open('/content/2024_state_of_the_union.txt') as file:
    files = file.read()
```

```python
text_split = CharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200,
    length_function=len
)
```

```python
chunks = text_split.create_documents([files])
```

```python
chunks[0]
```

```python
from langchain.embeddings import HuggingFaceEmbeddings

embedding_mode = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
VectorDB = Chroma(
    collection_name="Arquero",
    embedding_function=embedding_mode
)
```

```python
VectorDB
```

```python
storage_id = VectorDB.add_documents(chunks)
```

```python
len(storage_id)
```

```python
storage_id[0]
```

```python
result = VectorDB.similarity_search(
    query="what is joe biden talking about?",
    k=2
)
```

```python
result[0].page_content
```

```python
result[1].page_content
```

```python
retriver = VectorDB.as_retriever()
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-large")
```

```python
if tokenizer.pad_token is None:
    tokenizer.add_special_tokens({'pad_toke:[PAD]'})
```

```python
models = model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-large")
```

```python
model.resize_token_embeddings(len(tokenizer))
```

```python
model.config.pad_token_id = tokenizer.pad_token_id
```

```python
from transformers import pipeline

generator = pipeline(
    'text2text-generation',
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=150
)
```

```python
from langchain.llms import HuggingFacePipeline

llm = HuggingFacePipeline(pipeline=generator)
```

```python
template = """
use the context provided to answer the question. if you dont know the answer say you dont know.

context: {context}

question: {question}

answer:
"""
```

```python
custom_template = PromptTemplate(
    template=template,
)
```

```python
custom_template
```

```python
rag_chain = (
    {'context': retriver, 'question' :  RunnablePassthrough()}
    | custom_template
    | llm
    | StrOutputParser()
)
```

```python
query = 'what did  president say about Ukraine?'
answer = rag_chain.invoke(query)
```
