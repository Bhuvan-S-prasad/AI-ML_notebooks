```python
x = df_resampled.drop('machine_status',axis=1)
y = df_resampled['machine_status']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import StandardScaler

```

```python
sensor_cols = [col for col in x_train.columns if 'sensor' in col]

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train[sensor_cols])

vif_data = pd.DataFrame()
vif_data['Feature'] = sensor_cols
vif_data['VIF'] = [variance_inflation_factor(x_train_scaled, i) for i in range(len(sensor_cols))]

vif_data.sort_values(by='VIF', ascending=False, inplace=True)
print(vif_data)
```

```python
vif_threshold = 10

high_vif_features = vif_data[vif_data['VIF'] > vif_threshold]['Feature'].tolist()

X_train_reduced = x_train.drop(columns=high_vif_features)
X_test_reduced = x_test.drop(columns=high_vif_features)

print(f"Dropped due to VIF > {vif_threshold}: {high_vif_features}")
print(f"Remaining features: {X_train_reduced.columns.tolist()}")
```

```python
X_train_reduced
```

```python
corr = X_train_reduced.corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=False, cmap='coolwarm')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train_reduced, y_train)
```

```python
y_pred_lr = lr.predict(X_test_reduced)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test, y_pred_lr)*100)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(X_train_reduced, y_train)
```

```python
y_pred_dt = dt.predict(X_test_reduced)
```

```python
print(accuracy_score(y_test, y_pred_dt)*100)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train_reduced, y_train)
```

```python
y_pred_rf = rf.predict(X_test_reduced)
```

```python
print(accuracy_score(y_test, y_pred_rf)*100)
```
